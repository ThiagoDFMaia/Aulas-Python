# Papagaio

## Descrio
Os papagaios so aves coloridas e inteligentes, conhecidas por suas habilidades de imitao e vocalizao. Pertencem  famlia Psittacidae e so encontradas principalmente em regies tropicais e subtropicais.

## Caractersticas

- **Tamanho**: Variam de pequenos (como o periquito) a grandes (como a arara).
- **Cores**: Comumente apresentam plumagens vibrantes em verde, azul, amarelo e vermelho.
- **Bico**: Possuem um bico forte e curvado, ideal para quebrar sementes e frutos.

## Habitat
Os papagaios vivem em florestas tropicais, savanas e reas urbanas, dependendo da espcie. Muitas espcies so adaptveis e podem ser encontradas em reas modificadas pelo homem.

## Comportamento
- **Inteligncia**: So conhecidos por sua capacidade de aprender palavras e frases.
- **Socializao**: Vivem em grupos e tm um comportamento social complexo.
- **Alimentao**: A dieta  baseada em frutas, sementes, nozes e, ocasionalmente, insetos.

## Conservao
Muitas espcies de papagaios esto ameaadas devido  perda de habitat e ao trfico de animais. Programas de conservao so essenciais para a proteo dessas aves magnficas.

## Curiosidades
- Os papagaios podem viver mais de 50 anos em cativeiro.
- Algumas espcies, como o papagaio-do-mar, so famosas por suas habilidades de vocalizao.

## Concluso
Os papagaios so aves fascinantes que desempenham um papel importante nos ecossistemas onde habitam. A preservao de suas espcies e habitats  fundamental para a biodiversidade do planeta.


# Galinha

## Descrio
As galinhas so aves domsticas pertencentes  espcie Gallus gallus domesticus. Elas so amplamente conhecidas pela sua importncia na produo de ovos e carne, alm de serem populares como animais de estimao em diversas culturas.

## Caractersticas

- **Tamanho**: Geralmente variam de 1 a 3 kg, dependendo da raa.
- **Cores**: Podem apresentar plumagens em vrias cores, incluindo branco, marrom, preto e padronagens mescladas.
- **Bico**: Possuem um bico curto e forte, adaptado para ciscar no cho em busca de alimentos.

## Habitat
As galinhas so frequentemente mantidas em quintais, granjas e avirios. Elas preferem reas abertas com acesso a sombra e abrigo contra predadores.

## Comportamento
- **Socializao**: Galinhas so aves sociais e vivem em grupos chamados de bandos.
- **Comportamento de ciscar**: Passam grande parte do dia procurando alimento, como insetos e sementes, no solo.
- **Vocalizao**: Possuem um repertrio vocal variado para se comunicar entre si.

## Alimentao
A dieta das galinhas  variada e pode incluir gros, verduras, insetos e restos de comida.  importante que tenham acesso a uma alimentao balanceada para garantir sua sade e produtividade.

## Conservao
Embora as galinhas sejam uma das aves mais comuns no mundo, algumas raas esto ameaadas devido  produo industrial e  perda de diversidade gentica. Programas de conservao ajudam a preservar raas tradicionais.

## Curiosidades
- As galinhas podem reconhecer at 100 rostos diferentes.
- Elas possuem um ciclo de sono semelhante ao dos mamferos e podem sonhar.

## Concluso
As galinhas so aves fascinantes que desempenham um papel importante na agricultura e na cultura humana. A proteo das raas tradicionais e o bem-estar animal so essenciais para o futuro dessa espcie.

## Cria o modelo para treinamento

```python
from sklearn.svm import LinearSVC

modelo= LinearSVC()
```

## Treina o modelo com os dados e classes

```python
modelo.fit(dados,classes)
```

## Testes de previses dos animais

```python
animal=[1,1,0]
modelo.predict([animal])
```

```python
animal=[1,0,1]
modelo.predict([animal])
```

## Verifica a acuracia de acerto do modelo

```python
animal1=[1,1,1]
animal2=[1,1,0]
animal3=[0,1,1]
animal4=[1,0,1]
animal5=[0,0,1]
test_x=[animal1,animal2,animal3,animal4,animal5]

previsoes=modelo.predict(test_x)
```

```python
previsoes
```

```python
test_y=[1,0,1,1,1]
```

```python
from sklearn.metrics import accuracy_score

taxa_acerto= accuracy_score(test_y,previsoes)*100

print(f'Acuracia: {taxa_acerto}')
```
