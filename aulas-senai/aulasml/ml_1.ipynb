# Papagaio

## Descrição
Os papagaios são aves coloridas e inteligentes, conhecidas por suas habilidades de imitação e vocalização. Pertencem à família Psittacidae e são encontradas principalmente em regiões tropicais e subtropicais.

## Características

- **Tamanho**: Variam de pequenos (como o periquito) a grandes (como a arara).
- **Cores**: Comumente apresentam plumagens vibrantes em verde, azul, amarelo e vermelho.
- **Bico**: Possuem um bico forte e curvado, ideal para quebrar sementes e frutos.

## Habitat
Os papagaios vivem em florestas tropicais, savanas e áreas urbanas, dependendo da espécie. Muitas espécies são adaptáveis e podem ser encontradas em áreas modificadas pelo homem.

## Comportamento
- **Inteligência**: São conhecidos por sua capacidade de aprender palavras e frases.
- **Socialização**: Vivem em grupos e têm um comportamento social complexo.
- **Alimentação**: A dieta é baseada em frutas, sementes, nozes e, ocasionalmente, insetos.

## Conservação
Muitas espécies de papagaios estão ameaçadas devido à perda de habitat e ao tráfico de animais. Programas de conservação são essenciais para a proteção dessas aves magníficas.

## Curiosidades
- Os papagaios podem viver mais de 50 anos em cativeiro.
- Algumas espécies, como o papagaio-do-mar, são famosas por suas habilidades de vocalização.

## Conclusão
Os papagaios são aves fascinantes que desempenham um papel importante nos ecossistemas onde habitam. A preservação de suas espécies e habitats é fundamental para a biodiversidade do planeta.


# Galinha

## Descrição
As galinhas são aves domésticas pertencentes à espécie Gallus gallus domesticus. Elas são amplamente conhecidas pela sua importância na produção de ovos e carne, além de serem populares como animais de estimação em diversas culturas.

## Características

- **Tamanho**: Geralmente variam de 1 a 3 kg, dependendo da raça.
- **Cores**: Podem apresentar plumagens em várias cores, incluindo branco, marrom, preto e padronagens mescladas.
- **Bico**: Possuem um bico curto e forte, adaptado para ciscar no chão em busca de alimentos.

## Habitat
As galinhas são frequentemente mantidas em quintais, granjas e aviários. Elas preferem áreas abertas com acesso a sombra e abrigo contra predadores.

## Comportamento
- **Socialização**: Galinhas são aves sociais e vivem em grupos chamados de bandos.
- **Comportamento de ciscar**: Passam grande parte do dia procurando alimento, como insetos e sementes, no solo.
- **Vocalização**: Possuem um repertório vocal variado para se comunicar entre si.

## Alimentação
A dieta das galinhas é variada e pode incluir grãos, verduras, insetos e restos de comida. É importante que tenham acesso a uma alimentação balanceada para garantir sua saúde e produtividade.

## Conservação
Embora as galinhas sejam uma das aves mais comuns no mundo, algumas raças estão ameaçadas devido à produção industrial e à perda de diversidade genética. Programas de conservação ajudam a preservar raças tradicionais.

## Curiosidades
- As galinhas podem reconhecer até 100 rostos diferentes.
- Elas possuem um ciclo de sono semelhante ao dos mamíferos e podem sonhar.

## Conclusão
As galinhas são aves fascinantes que desempenham um papel importante na agricultura e na cultura humana. A proteção das raças tradicionais e o bem-estar animal são essenciais para o futuro dessa espécie.

## Cria o modelo para treinamento

```python
from sklearn.svm import LinearSVC

modelo= LinearSVC()
```

## Treina o modelo com os dados e classes

```python
modelo.fit(dados,classes)
```

## Testes de previsões dos animais

```python
animal=[1,1,0]
modelo.predict([animal])
```

```python
animal=[1,0,1]
modelo.predict([animal])
```

## Verifica a acuracia de acerto do modelo

```python
animal1=[1,1,1]
animal2=[1,1,0]
animal3=[0,1,1]
animal4=[1,0,1]
animal5=[0,0,1]
test_x=[animal1,animal2,animal3,animal4,animal5]

previsoes=modelo.predict(test_x)
```

```python
previsoes
```

```python
test_y=[1,0,1,1,1]
```

```python
from sklearn.metrics import accuracy_score

taxa_acerto= accuracy_score(test_y,previsoes)*100

print(f'Acuracia: {taxa_acerto}')
```
