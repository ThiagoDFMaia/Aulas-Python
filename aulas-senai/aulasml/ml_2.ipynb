# Abrindo a planilha com os dados

```python
import pandas as pd
dados= pd.read_csv('dataset/tracking.csv')
dados.head(5)
```

```python
comprou=dados.comprou
dados.drop(columns='comprou',inplace=True)
comprou.head()
```

```python
dados.columns
```

```python
from sklearn.model_selection import train_test_split


treino, teste, classe_treino, classe_teste = train_test_split(dados, comprou,random_state=42, test_size=0.3)
```

```python
dados.info()
```

```python
teste.info()
```

```python
treino.info()
```

```python
classe_treino.info()
```

```python
classe_teste.info()
```

```python
from sklearn.svm import LinearSVC

modelo= LinearSVC()
```

```python

modelo.fit(treino,classe_treino)
```

```python
teste.head()
```

```python
classe_teste.head()
```

```python
t=[1,1,1,0]
modelo.predict([t])
```

```python
t=[0,1,1,0]
modelo.predict([t])
```

# Verificando a acurácia

```python
teste
```

```python
previsoes=modelo.predict(teste)
```

```python
previsoes
```

```python
from sklearn.metrics import accuracy_score

taxa_acerto= accuracy_score(classe_teste,previsoes)*100
print(f'Acuracia: {taxa_acerto}')
```
